## Pré-processamento de dados.
Problemática: Cartão de Crédito

```python
#Chamando bibliotecas 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cc = pd.read_csv('/Users/tulioteodoro/Documents/GitProjects/credit_card/data/CC_DATASET.csv')
```

```python
cc.head()
```

```python
cc['PURCHASES'].max()
```

```python
for i in cc.drop(['CUST_ID'], axis=1):
    cc[i] = np.where(cc[i] <= cc[i].quantile(0.0001),
                     
                    cc[(cc[i] > cc[i].quantile(0.001)) &
                       (cc[i] < cc[i].quantile(0.999))][i].mean(),
                     cc[i])
    cc[i] = np.where(cc[i] >= cc[i].quantile(0.999),
                    
                    cc[(cc[i] > cc[i].quantile(0.001)) &
                       (cc[i] < cc[i].quantile(0.999))][i].mean(),
                     cc[i])
```

```python
cc['PURCHASES'].max()

#A escolha do loop for utilizando os dados que estão entre 1% e 99% já remove os outliers necessários 
```

## Missing values

```python
cc.fillna(value=0, inplace=True)
cc.head()
```

## Normalização

```python
from sklearn.preprocessing import StandardScaler
estandcc = StandardScaler()
```

```python
#Preparando os dados para o FIT, excluir a coluna CUST_ID já que os dados não serão utilizados.
estandcc.fit(cc.drop(['CUST_ID'],axis=1))
```

```python
#Transformando os dados do seguro auto em dados normalizados.
valestandcc = estandcc.transform(cc.drop(['CUST_ID'], axis=1))
```

```python
#O Objetor valestandcc criado é do tipo np array, agora criaremos um DF a partir do array.

dfccestand = pd.DataFrame(valestandcc, columns = cc.drop(['CUST_ID'],axis=1).columns)
dfccestand.head()
```

## Modelagem

Clusterização sobre os dados normalizados.

```python
# Importando o k-means do scikit-learn
from sklearn.cluster import KMeans
```

```python
#Definimos em k a quantidade de grupos que queremos "dividr os dados"
k2 = 2 

#Instanciando o objeto k-means, como serão duas clusterizações, definimos k2 e k3
modeloKM_K2 = KMeans(n_clusters= k2)

#Fitanod o modelo usando os dados normalizados
modeloKM_K2.fit(X = dfccestand)

#Adicionando ao df os clusters, a partir da propriedade de labels do objeto k-means modeloKM_K2
cc['CLUSTER_K2'] = pd.DataFrame(modeloKM_K2.labels_).values
```

```python
cc.head()
```

```python
k3 = 3
modeloKM_K3 = KMeans(n_clusters= k3)
```

```python
modeloKM_K3.fit(X = dfccestand)
cc['CLUSTER_K3'] = pd.DataFrame(modeloKM_K3.labels_).values
```

```python
cc.head()
```

## Análise de dados

1. Escolhendo duas métricas para plotar o gráfico de dispersão.

```python
#PLOT
plt.figure(figsize=(20,10))
sns.lmplot(x = 'PAYMENTS', y = 'PURCHASES', data = cc, hue = 'CLUSTER_K2', fit_reg=False)
```

```python
#Plotando figura com cluster k=3

plt.figure(figsize=(15,8))
sns.lmplot(x = 'PAYMENTS', y = 'PURCHASES', data = cc, hue = 'CLUSTER_K3', fit_reg=False)
```

CASOS QUE FICARAM ENQUADRADOS EM CADA CLUSTER EM CADA UMA DAS CLUSTERIZAÇÕES

```python
print(cc.groupby('CLUSTER_K2')['CUST_ID'].count())
print('')
print(cc.groupby('CLUSTER_K3')['CUST_ID'].count())
```

## A média e a mediana de cada cluster em cada uma das clusterizações

```python
print('MÉDIA DE COMPRAS')
print(cc.groupby('CLUSTER_K2')['PURCHASES'].mean())
print('')
print(cc.groupby('CLUSTER_K3')['PURCHASES'].mean())
print('')
print('MEDIANA DE COMPRAS')
print(cc.groupby('CLUSTER_K2')['PURCHASES'].median())
print('')
print(cc.groupby('CLUSTER_K3')['PURCHASES'].median())
```

## Usando a clusterização k=3 para descobrir a mediana do limite de crédito de cada grupo

```python
cc.groupby('CLUSTER_K3')['CREDIT_LIMIT'].median()
```
